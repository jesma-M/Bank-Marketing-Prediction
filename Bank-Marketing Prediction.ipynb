```python
numerical_features=['age','campaign','duration']
for cols in numerical_features:
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1     

    filters = (df[cols] >= Q1 - 1.5 * IQR) & (df[cols] <= Q3 + 1.5 *IQR)
    df=df.loc[filters]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
def tuning(clf,parameter,X_train,y_train):
    cv=StratifiedKFold(n_splits=5,shuffle=True)
    cls_grid=GridSearchCV(clf,parameter,cv=cv,scoring='f1',n_jobs=-1)
    cls_grid.fit(X_train,y_train)
    best_hp=cls_grid.best_params_
    return cls_grid.best_estimator_,best_hp
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
ac=accuracy_score(y_test,y_pred)
```

```python
ac
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
dt=DecisionTreeClassifier()
```

```python
dt_parameter_grid={'criterion':['gini','entropy'],
                  'max_depth':[12,13,14]}
```

```python
tuning(dt,dt_parameter_grid,X_train,y_train)
```
